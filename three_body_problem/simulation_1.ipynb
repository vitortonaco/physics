```python
def Fg(ma,mb,Xa,Xb):
    # a e b devem ser numeros entre 1,2,3

     
    
    rba = Xb-Xa
    d = norm(rba)
    
    
    fx = (G*ma*mb*(rba[0])/d**3)
    fy = (G*ma*mb*((rba[1])/d**3))
    fz = (G*ma*mb*((rba[2])/d**3))
    
          
    return np.array([fx,fy,fz])
    
    
norm = lambda x : (np.sqrt(x[0]**2+x[1]**2+x[2]**2))
    
```

```python
M0 = 5.974*10**24
V = 107000/3.6 #m/s

G = 6.6743*10**(-11)

Ms = 1.98892*10**30 #kg
# R = 384400000 # km
D = 149600000000 # Distance between the sun and Earth
Fc = M0*V**2/D # Centripetal force
years = 3
tf = years*365*24*60*60


hora = 3600 # s

# V = R/(hour)

cos = np.cos
sin = np.sin

m = [Ms,M0,M0] # kg
m1,m2,m3 = m[0],m[1],m[2]


theta = 2*math.pi/3



v1 = [0.3*V,0,V/100]
v2 = [V*np.sin(theta),V*np.cos(theta),0]
v3 = [V*np.sin(theta),-V*np.cos(theta),0]


X1 = np.array([0,0,0])
X2 = np.array([D*cos(theta+math.pi/2),D*sin(theta+math.pi/2),0])
X3 = np.array([D*cos(theta-5*math.pi/2),-D*sin(theta-5*math.pi/2),0])




x0 = [X1[0],v1[0],X1[1],v1[1],X1[2],v1[2],
      X2[0],v2[0],X2[1],v2[1],X2[2],v2[2],
      X3[0],v3[0],X3[1],v3[1],X3[2],v3[2]]


# n = Normalizer().fit(np.arra)
# x0 = n.transform(x0)[0]
```

```python
def newtons_eq(x,t):
    
    
    
    X1 = np.array([x[0],x[2],x[4]])
    X2 = np.array([x[6],x[8],x[10]])
    X3 = np.array([x[12],x[14],x[16]])
    
    
    d12 = norm(X1-X2)
    d13 = norm(X1-X3)
    d23 = norm(X3-X2)
    
    m1,m2,m3 = m[0],m[1],m[2]
    
    # m1 ////////////////////////////////////////////////////////////////////////////////////////////
    
    F12 = Fg(m1,m2,X1,X2)
    F13 = Fg(m1,m3,X1,X3)
    
    xdot0 = x[1]
    xdot1 = (F12[0]+F13[0])/m[0]
    
    xdot2 = x[3]
    xdot3 = (F12[1]+F13[1])/m[0]
    
    xdot4 = x[5]
    xdot5 = (F12[2]+F13[2])/m[0]
    
    # m2 ////////////////////////////////////////////////////////////////////////////////////////////
    F21 = -F12
    F23 = Fg(m2,m3,X2,X3)
    
    
    xdot6 = x[7]
    xdot7 = (F21[0]+F23[0])/m[1]
    
    xdot8 = x[9]
    xdot9 = (F21[1]+F23[1])/m[1]
    
    xdot10 = x[11]
    xdot11 = (F21[2]+F23[2])/m[1]
    
    # m3 ////////////////////////////////////////////////////////////////////////////////////////////
    
    F31 = - F13
    F32 = -F23
    
    
    xdot12 = x[13]
    xdot13 = (F31[0]+F32[0])/m[2]
    
    xdot14 = x[15]
    xdot15 = (F31[1]+F32[1])/m[2]
    
    xdot16 = x[17]
    xdot17 = (F31[2]+F32[2])/m[2]
    
    
    
    
            
    return [xdot0,xdot1,xdot2,xdot3,xdot4,xdot5,
            xdot6,xdot7,xdot8,xdot9,xdot10,xdot11,
            xdot12,xdot13,xdot14,xdot15,xdot16,xdot17]

t = np.linspace(0,tf,years*365)

newtons_eq(x0,t)
    
    
    
answer = odeint(newtons_eq,x0,t)

```
