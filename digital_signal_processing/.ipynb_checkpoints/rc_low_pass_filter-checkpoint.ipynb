## Transfer Function

A simple one pole RC filter has the following transfer function

$$
G(s) = \frac{1}{1+\tau\cdot{s}}
$$

And the following differential equation

$$
V_{out}(t) = V_{in}(t)-RC \frac{dV_{out}}{dt}
$$

```python
# Constants and functions

hz2rad = lambda f: 2*math.pi*f
rad2hz = lambda omega: omega/(2*math.pi)
fc = hz2rad(100)
tau = 1/fc # Time constant
F = 500 #Hz
t = np.linspace(0,1,44100) # 44.1 khz of sample rate
dt = t[1]-t[0]
freqs = np.fft.fftfreq(len(t),dt)

def H_RC(omega):
    s = omega*1j
    den = 1 + tau*s
    return 1/den


def H_ChebyshevBand(freq, fc, stageCount):
    C = 0.01*10**(-6) # Faraday
    R1 = 10*10**3 # ohm
    stages = [(0.5359, 0.522), (0.2657, 1.379),
              (0.1848, 1.711), (0.1582, 1.913)] # pairs of (k1,k2)
    
      
    s = 1j * (freq)
    G = 1
    
    for i in range(stageCount):
        stage = stages[i]
        R = stage[0]/(C*fc)
        Rf = R1*stage[1]  
        G *=(4*R*C*(R1+Rf)*s)/(8*(R**2)*(C**2)*(s**2)*(R1-Rf)+2*R*C*(3*R1-Rf)*s+R1) 
    
    return np.abs(G)


def H_ChebyshevLow(freq, fc, stageCount):
    C = 0.01*10**(-6) # Faraday
    R1 = 10*10**3 # ohm
    stages = [(0.5359, 0.522), (0.2657, 1.379),
              (0.1848, 1.711), (0.1582, 1.913)] # pairs of (k1,k2)
    w = hz2rad(freq)
    if (freq == 0):
        return 0
      
    s = 1j * (w)
    G = 1
    
    for i in range(stageCount):
        stage = stages[i]
        R = stage[0]/(C*fc)
        Rf = R1*stage[1]
        Zc = -1j/(C*w) # impedancia do capacitor
        
        k1 = (2*Zc+R)/(3*R*Zc + R*R + Zc*Zc)
        k2 = (Zc+R)/(3*R*Zc + R*R + Zc*Zc)
        num = 1 - 2*k1*R + R*(1-R*k1)/Zc
        den = R1/(R1+Rf) - 2*R*k2 + R*(1-R*k2)/Zc
        G *= num/den
    
    return G

def Phase(H_w):
    
    return math.atan(np.imag(H_w)/np.real(H_w))

def SawWave(x,f):
    amp = 3
    T = 1/f
    
    dydx = (-2*amp)/T 
    return (x%T)*dydx + amp

def SineWave(x,f):
    amp = 3
    T = 1/f
    w = hz2rad(f)
    
    return amp*np.sin(x*w)

def SquareWave(xIn,f):
    amp = 3
    T = 1/f
    if (xIn%T > 0.5*T):
        return amp
    else: 
        return -amp

    
def plotBode(w, H, cutoffFreq, logScale = True):
    fig,axs = plt.subplots(2,1)
    G = np.abs(H)# gain
    vectorizedPhase = np.vectorize(Phase)
    P =  vectorizedPhase(H)# phase
    axs[0].plot(w[:len(freqs)//2],G[:len(freqs)//2])
    axs[0].axvline(x = cutoffFreq, ls = '--', c='black', lw=2)
    axs[0].set_title('Gain plot (cutoff = {:.2f}Hz)'.format(rad2hz(cutoffFreq)))
    axs[0].set_ylabel('$|G(j\omega)|$')
    if (logScale):
        axs[0].set_yscale('log')
        axs[0].set_xscale('log')
    axs[0].set_xlim(0.1, np.max(w))
    axs[0].set_ylim(10**(-5),10-10**(-10))

    axs[1].plot(w[:len(freqs)//2],P[:len(freqs)//2])
    axs[1].axvline(x = cutoffFreq, ls = '--', c='black', lw=2)
    if (logScale):
        axs[1].set_xscale('log')
    axs[1].set_xlim(0.1, np.max(w))


    axs[1].set_title('Phase plot')
    axs[1].set_ylabel('$Phase(\omega)$')
    axs[1].set_xlabel('$\omega$')
    axs[1].grid()
    #axs[1].set_ylim(-0.02,0.02)

    axs[0].grid()
    fig.set_size_inches(12,8)
    return axs, fig
```

# RC Filter

```python
H_w = H_RC(freqs)

################################ plot
H_w
plotBode(freqs,H_w,fc)
```

# Aliasing analysis

    Aliasing is the effect of an analog signal being sampled to a digital one and getting distorted in this process depending on the frequency rate of the process. A low sample rate will lead to a misleading representation of the waveform and this can be noticed not only in its shape, but in its frequency spectrum. By the Nyquist theorem, the sample rate shoulde be at least the double of the highest frequency you are trying to represent in a digital system.
    
    As you can see in the graphs, even though you achive double the frequency you one to represent, this does not mean that your signal is safe from distortion, it only means that now your digital signal contains the analog main frequency, but it still got distorted by aliasing

```python
def get_wave_from_sample_rate(sampleRate, t0, tf, Fsignal):
    dt = 1/sampleRate
    size = (tf-t0)//dt + 1
    t = np.linspace(t0,tf,int(size))
    signal = [SawWave(i,Fsignal) for i in t]
    fft = np.fft.fft(signal)
    omegas = np.fft.fftfreq(int(size),dt)
    return t,signal, np.abs(fft[:len(omegas)//2]), omegas[:len(omegas)//2]
    
```

```python
t0 = 0
tf = 4
fs = 44.1*10**3  #sample rate

Fsignal = 1000 # hz

fig, axs = plt.subplots(8,2)
sampleRates = np.logspace(3.1,4.2,8)

for i in range(8):
    fs = sampleRates[i]
    t2, x2 , fft, omegas = get_wave_from_sample_rate(fs, t0, tf, Fsignal)
    axs[i,0].plot(t2,x2)
    axs[i,0].set_xlabel('time [s]')
    axs[i,0].set_xlim(0,6/Fsignal)
    axs[i,0].set_title('{:.2f} kHz sample rate'.format(fs/1000))
    axs[i,1].set_title('Frequency spectrum')
    axs[i,1].plot(omegas,fft)
    axs[i,1].set_yscale('log')
    axs[i,1].set_xscale('log')
    axs[i,1].set_xlim(20,20000)
    axs[i,1].set_ylim(10**2,10**6)
    
    
    
fig.set_size_inches(17,45)
```
